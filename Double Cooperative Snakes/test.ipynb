```python
import gym
import gym
import numpy as np
import torch

from snake_double import SnakeMultiAgentEnv

from mppo import PPOAgent

import pygame
# Initialize Pygame
pygame.init()
pygame.font.init()
torch.set_num_threads(4)
```

```python
HEIGHT = 10
WIDTH = 10
env = SnakeMultiAgentEnv(width=600, height=600, rows=HEIGHT, cols=HEIGHT)
agent = PPOAgent(height=HEIGHT,width=WIDTH)
checkpoint = 'ppo_cnn_double_snake_checkpoint.pth'
```

```python
try:
    agent.load(checkpoint)
except Exception as e:
    print(e)
```

```python
# Train the agent
agent.train(env, num_episodes=5000,checkpoint_path=checkpoint)
```

```python
agent.plot("ppo_double_cnn_snake999.png")
```

```python
agent.test(env)
```
